```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as mplt
import seaborn as sn
%matplotlib inline
```

```python
df23 = pd.read_csv("../data/23_anti_corruprion_cases.csv")
df24 = pd.read_csv("../data/24_anti_corruption_arrests.csv")
```

```python
df23.shape, df24.shape
```

There isn't any records of anti corruption cases filed in Uttarkhand in 2003

```python
df24.head()
```

```python
df = df23.merge(
    df24,
    on=['state_ut', 'year'],
    how='outer'
)
```

```python
yr_grp = df.groupby('year').aggregate(sum).reset_index()
```

```python
yr_grp.head()
```

```python
sn.set_style('darkgrid')
fig, ax = mplt.subplots()
fig.set_size_inches(15, 8)
x_axis1 = yr_grp[['cases_registered', 'cases_pending_investigation_from_previous_year']]\
    .sum(axis=1)\
    .map(lambda x: x / 1000)
value_markers = yr_grp[
    [ 'persons_convicted_during_the_year']
].sum(axis=1)
ax.scatter(
    yr_grp.year,
    x_axis1,
    c='r'
)
for x, y, z in zip(yr_grp.year, x_axis1, value_markers):
    ax.annotate(str(z) + ' convicts', xy=(x - 0.07,y + 0.05))
ax.set_title("Anti Corruption cases dealt from 2001 - 2010 (in thousands) and no. of convicts")
ax.set_xlabel('Year')
ax.set_ylabel('Count')
```

```python
columns_of_interest = [
    'state_ut',
    'year',
    'cases_pending_investigation_from_previous_year',
    'cases_registered',
    'total_cases_for_investigation',
    'no_of_cases_investigated',
    'cases_pending_departmental_sanction_for_prosecution',
    'cases_pending_investigation_at_the_end_of_the_year',
    'cases_pending_trial_from_the_previous_year',
    'total_cases_for_trial',
    'cases_convicted',
    'total_amount_of_fine_imposed_in_rs',
    'persons_in_custody_or_on_bail_during_the_stage_of_investigation_at_the_beginning_of_the_year',
    'persons_arrested_during_the_year',
    'persons_in_custody_or_on_bail_during_the_stage_of_investigation_at_the_end_of_the_year',
    'persons_under_trial_at_the_beginning_of_the_year',
    'persons_convicted_during_the_year',
    'persons_punished_departmentally_during_the_year',
    'no_of_group_a_officers_out_of_above',
    'no_of_group_b_officers_out_of_above',
    'private_persons_involved_during_the_year'
]
```
